```python
import pandas as pd
import numpy as np
import os

print("Files:", os.listdir())

file_name = "Most Popular Programming Languages.csv"
df = pd.read_csv(file_name)

df.columns = df.columns.str.strip()
df['Month'] = pd.to_datetime(df['Month'])

language_columns = sorted([col for col in df.columns if col != 'Month'])

student_id = "2025-1136"
last_three_digits = int(student_id[-3:])
index = last_three_digits % len(language_columns)

assigned_language = language_columns[index]
print("Assigned Language:", assigned_language)

data = df[['Month', assigned_language]].copy()
data.rename(columns={assigned_language: 'Popularity'}, inplace=True)

data = data.sort_values('Month')

data['Growth_Rate'] = data['Popularity'].pct_change() * 100

window = 6
data['Moving_Avg'] = data['Popularity'].rolling(window=window).mean()
data['Moving_STD'] = data['Popularity'].rolling(window=window).std()

mean_growth = np.mean(data['Growth_Rate'].dropna())
std_growth = np.std(data['Growth_Rate'].dropna())

print("\nMean Growth Rate:", mean_growth)
print("STD Growth Rate:", std_growth)

conditions = [
    (data['Growth_Rate'] > mean_growth),
    (data['Growth_Rate'] > 1) & (data['Growth_Rate'] <= mean_growth),
    (data['Growth_Rate'].between(-1, 1)),
    (data['Growth_Rate'] < -std_growth)
]

choices = ['Growth', 'Introduction', 'Maturity', 'Decline']

data['Lifecycle_Phase'] = np.select(conditions, choices, default='Maturity')

phase_counts = data['Lifecycle_Phase'].value_counts()
phase_percent = (phase_counts / len(data)) * 100

dominant_phase = phase_counts.idxmax()

print("\n--- Lifecycle Table Sample ---")
print(data.head(10))

print("\n--- Lifecycle Counts ---")
print(phase_counts)

print("\n--- Lifecycle Percentages (%) ---")
print(phase_percent.round(2))

print("\nDominant Lifecycle Stage:", dominant_phase)

print("\n--- Interpretation ---")
if dominant_phase == 'Growth':
    print(f"{assigned_language} is in a strong growth phase with increasing adoption.")
elif dominant_phase == 'Introduction':
    print(f"{assigned_language} is emerging with gradual adoption.")
elif dominant_phase == 'Maturity':
    print(f"{assigned_language} is in a maturity phase with stable usage.")
elif dominant_phase == 'Decline':
    print(f"{assigned_language} shows signs of decline in popularity.")
```

```python
import pandas as pd
import numpy as np
import os

# === 0. CHECK FILE ===
print("Files:", os.listdir())

# === 1. LOAD DATA ===
file_name = "Most Popular Programming Languages.csv"
df = pd.read_csv(file_name)

# Clean column names
df.columns = df.columns.str.strip()

# Convert Month
df['Month'] = pd.to_datetime(df['Month'])

# === 2. SORT LANGUAGE COLUMNS ===
language_columns = sorted([col for col in df.columns if col != 'Month'])

# === 3. ASSIGN LANGUAGES ===
student_id = "2025-1136"  # 🔁 change this
last_three_digits = int(student_id[-3:])

index = last_three_digits % len(language_columns)

lang_A = language_columns[index]

# Next language alphabetically
if index + 1 < len(language_columns):
    lang_B = language_columns[index + 1]
else:
    lang_B = language_columns[0]  # wrap around

print("Language A:", lang_A)
print("Language B:", lang_B)

# === 4. EXTRACT DATA ===
data = df[['Month', lang_A, lang_B]].copy()
data.rename(columns={lang_A: 'Lang_A', lang_B: 'Lang_B'}, inplace=True)

# Sort by time
data = data.sort_values('Month')

# === 5. CENTRAL TENDENCIES ===
mean_A = data['Lang_A'].mean()
median_A = data['Lang_A'].median()
std_A = data['Lang_A'].std()

mean_B = data['Lang_B'].mean()
median_B = data['Lang_B'].median()
std_B = data['Lang_B'].std()

# === 6. NORMALIZATION (Min-Max Scaling) ===
norm_A = (data['Lang_A'] - data['Lang_A'].min()) / (data['Lang_A'].max() - data['Lang_A'].min())
norm_B = (data['Lang_B'] - data['Lang_B'].min()) / (data['Lang_B'].max() - data['Lang_B'].min())

# === 7. CORRELATION ===
correlation = np.corrcoef(data['Lang_A'], data['Lang_B'])[0, 1]

# === 8. DOMINANCE RATIO ===
dominance_count = (data['Lang_A'] > data['Lang_B']).sum()
total_months = len(data)

dominance_ratio = (dominance_count / total_months) * 100

# === 9. RELATIVE PERFORMANCE INDEX (RPI) ===
rpi = np.mean(norm_A - norm_B)

# === 10. CROSSOVER POINTS ===
data['A_gt_B'] = data['Lang_A'] > data['Lang_B']

# Detect changes (False → True means crossover)
crossover_points = data[data['A_gt_B'].diff() == True]

# === 11. SUMMARY TABLE ===
summary = pd.DataFrame({
    'Mean_A': [mean_A],
    'Mean_B': [mean_B],
    'Std_A': [std_A],
    'Std_B': [std_B],
    'Correlation': [correlation],
    'Dominance_Ratio (%)': [dominance_ratio],
    'RPI': [rpi]
})

# === 12. OUTPUT ===
print("\n--- Data Sample ---")
print(data.head(10))

print("\n--- Summary Statistics ---")
print(summary)

print("\n--- Dominance ---")
print(f"Language A dominated {dominance_count} out of {total_months} months")

print("\n--- Crossover Points ---")
print(crossover_points[['Month', 'Lang_A', 'Lang_B']])

# === 13. INTERPRETATION ===
print("\n--- Interpretation ---")

if dominance_ratio > 60:
    print(f"{lang_A} is the dominant language over time.")
elif dominance_ratio < 40:
    print(f"{lang_B} is the dominant language over time.")
else:
    print("Both languages show competitive performance.")

if std_A < std_B:
    print(f"{lang_A} is more stable (less volatile).")
else:
    print(f"{lang_B} is more stable (less volatile).")

if correlation > 0.7:
    print("The languages have strong positive correlation (similar trends).")
elif correlation < -0.7:
    print("The languages have strong inverse relationship.")
else:
    print("The languages show weak/moderate relationship.")
```
